```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
```

```python
# x = number of atoms 
# y = number of temp steps used in this work for a ~ 290-450 K temperature range 
# (AAQAA3, fragment ACTR, ACTR, HTTex1 16Q)
x = [29730,47040,95440,174602]
y = [48, 70, 80, 100]
```

```python
# expected to scale linearly with sqrt(N)
newx = np.sqrt(x)
newy = y

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(newx, newy)

print("slope:", slope)
print("intercept:", intercept)
print("r:", r_value)
print("r^2:", r_value**2)
print("p-value:", p_value)
print("std_err:", std_err)
```

```python
plt.scatter(newx, newy)

fitx = np.sqrt(np.arange(10000, 300000))
fity = fitx*slope + intercept
plt.plot(fitx, fity)
plt.xlabel('sqrt(N)')
plt.ylabel('Number of temp. steps')


```
