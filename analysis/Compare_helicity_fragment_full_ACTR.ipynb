```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import pyblock
from scipy.special import logsumexp
from scipy.special import softmax

# matplotlib settings
# --- Publication style settings ---
plt.rcParams.update({
    # Font
    "font.family": "Times New Roman",
    "font.serif": ["Arial"],  # or other preferred serif font
    "mathtext.fontset": "cm",            # Computer Modern for math text
    "text.usetex": False,                 # Use LaTeX for text rendering]
    "pdf.fonttype": 42,                 # Use LaTeX for text rendering

    # Figure
    "figure.figsize": (6, 4),            # width, height in inches
    "figure.dpi": 300,                   # high resolution

    # Axes
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "axes.linewidth": 1.2,
    "axes.grid": True,                   # optional, remove if you want no grid
    "grid.linestyle": "--",
    "grid.alpha": 0.7,

    # Ticks
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.top": True,
    "ytick.right": True,

    # Lines
    "lines.linewidth": 2,
    "lines.markersize": 6,

    # Legend
    "legend.fontsize": 10,
    "legend.frameon": False
})


output_path = './plots_ACTR_fragment_vs_full/'
prod_name = '300K'
```

```python
# Helicity at 300 K

# save data
dffrag= pd.read_csv('plots_fragmentACTR/300K_perres_helicity_300K.csv')
dffull= pd.read_csv('plots_ACTR/300K_perres_helicity_300K.csv')

# plot PMF with blocking error bars
plt.figure(figsize = (4,3))
plt.title(f"300 K")
plt.errorbar(dffrag['Residue'], dffrag['Fraction helix'], yerr=dffrag[''], fmt='o-', markersize=5, capsize=2, linewidth=1., elinewidth=1., color='#779AEC', label = 'ACTR$_{20-60}$')
plt.errorbar(dffull['Residue'], dffull['Fraction helix'], yerr=dffull[''], fmt='o-', markersize=5, capsize=2, linewidth=1., elinewidth=1., color='purple', label = 'ACTR')
plt.xlabel("Residue")
plt.ylabel("Fraction helix")
plt.ylim(0,1.)
plt.legend()
plt.tight_layout()
plt.savefig(output_path+f"{prod_name}_perres_helicity_300K.pdf")
```
